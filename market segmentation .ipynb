## Market segmentation based on Geographical feature as Segmentation Variable

#### combine dataset along states

```python
ev_state_merged = ev_state.merge(ev_charging_22 ,on ='State/UT' , how = 'left')
```

```python
ev_state_merged = ev_state_merged.fillna(0)
ev_state_merged
```

```python
ev_state_merged.drop('State/UT',axis = 1 , inplace = True)
```

## K-mean Clustering algorithm  

```python
# feature scaling 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
ev_state_merged = scaler.fit_transform(ev_state_merged)
```

```python
ev_state_merged
```

```python
pd.DataFrame(ev_state_merged).describe()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans()
```

```python
kmeans = KMeans(n_clusters = 2 , init = 'k-means++')
# fitting the kmeans algorithm on scaled data 
kmeans.fit(ev_state_merged)
```

```python
kmeans.inertia_
```

```python
# elbow method to select k value 
wcss = []
for k in range(1,8):
  kmeans = KMeans(n_clusters = k ,init = 'k-means++')
  kmeans.fit(ev_state_merged)
  wcss.append(kmeans.inertia_)
wcss
```

```python
# conversion the results into dataframe and plotting them
df = pd.DataFrame({'Cluster' : range(1,8) ,'WCSS': wcss})
plt.figure(figsize =(16,8))
plt.plot(df['Cluster'],df['WCSS'] , marker = 'o')
plt.title("elbow curve")
plt.xlabel("number of clusters")
plt.ylabel("within cluster squared ")
```

```python
means = KMeans(n_clusters = 5 , init = 'k-means++')
kmeans.fit(ev_state_merged)
pred = kmeans.predict(ev_state_merged)
```

```python
pred
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)

df_pca = pd.DataFrame(pca.fit_transform(ev_state_merged))

df_pca["cluster"] = pred
df_pca
```

```python
centers = kmeans.cluster_centers_
centers = pca.fit(centers)
```

## plotting clusters

```python
sns.scatterplot(x =0 , y=1,hue ="cluster" ,data = df_pca,palette = "bright")
plt.title("Clusters")
```
