```python
# scaling the values 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaled_values = pd.DataFrame(scaler.fit_transform(df1) , columns = df1.columns)
```

```python
# applying PCA 
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
```

```python
pca_data = pca.fit_transform(scaled_values)
```

```python
pca_data
```

```python
pca_data = pd.DataFrame(pca_data , columns = ["PCA1" , "PCA2"])
pca_data
```

## K-Means clustering algorithm 

```python
from sklearn.cluster import KMeans
```

```python
wcss = [] 
for i in range(1,11):
 kmeans = KMeans(n_clusters = i , init = "k-means++" , random_state = 30)
 kmeans.fit(scaled_values)
 wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
# plotting the kmeans graph between k and wcss
range_val = range(1,11)
plt.plot(range_val,wcss, marker = "o")
plt.xlabel('the number of cluster ')
plt.xticks(np.arange(1,11,1))
plt.ylabel('Inertia')
plt.title('Scree/ Elbow Plot')
plt.show()
```

```python
wcss = [] 
for i in range(1,11):
 kmeans = KMeans(n_clusters = i , init = "k-means++" , random_state = 30)
 kmeans.fit(df1)
 wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),wcss,marker="o")
plt.xlabel("k values")
plt.xticks(np.arange(1,11,1))
plt.ylabel("Inertia")
plt.title("Elbow Plot")
plt.show()
```

### we get k value = 3

```python
model = KMeans(4)
model.fit_predict(pca_data)
pca_df_kmeans = pd.concat([pca_data, pd.DataFrame({'cluster': model.labels_})], axis = 1)
```
